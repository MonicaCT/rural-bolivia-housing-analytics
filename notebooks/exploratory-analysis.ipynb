# Guided exploratory analysis

**Public-data status:** all observations are synthetic. This notebook is for exploration; the production pipeline lives in `src/`.

```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
households = pd.read_csv(ROOT / 'data/synthetic/households_synthetic.csv')
members = pd.read_csv(ROOT / 'data/synthetic/members_synthetic.csv')
households.shape, members.shape
```

## Start with denominators and missingness
A rigorous analysis states the unit of analysis and missing denominator before reporting percentages.

```python
missing = households.isna().mean().sort_values(ascending=False).head(10)
missing.to_frame('missing_share')
```

## Examine distributions, not only averages

```python
sns.displot(households, x='monthly_income_bob', hue='head_gender', kind='ecdf')
plt.xlabel('Monthly household income (BOB)')
plt.show()
```

## Avoid causal overstatement
The following table describes associations. It cannot establish that education or income causes housing adequacy.

```python
households.groupby('adequate_housing')[['head_education_years', 'monthly_income_bob', 'housing_score']].median()
```
